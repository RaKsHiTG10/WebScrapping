```python
# TASK 3: Calculate the distances between a launch site to its proximities
import math

# Haversine distance function
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in KM
    lat1, lon1, lat2, lon2 = map(math.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = math.sin(dlat/2)**2 + math.cos(lat1)*math.cos(lat2)*(math.sin(dlon/2)**2)
    c = 2 * math.asin(math.sqrt(a))
    
    return R * c
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.plugins import MarkerCluster
import pandas as pd
from math import radians, cos, sin, asin, sqrt

# ------------------------------------------
# Haversine Distance Function
# ------------------------------------------
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # km
    dlat = radians(lat2 - lat1)
    dlon = radians(lon2 - lon1)
    lat1 = radians(lat1)
    lat2 = radians(lat2)

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    return 2 * R * asin(sqrt(a))

# ------------------------------------------
# Example Launch Site (Use real DF values)
# ------------------------------------------
launch_site = {
    "name": "CCAFS LC-40",
    "lat": 28.562302,
    "lon": -80.577356
}

# Example proximity points
cities = [
    ("Titusville", 28.6123, -80.8075),
    ("Cape Canaveral", 28.3922, -80.6077),
    ("Orlando", 28.5383, -81.3792)
]

railways = [
    ("Railway Depot", 28.5830, -80.8000)
]

highways = [
    ("Highway 1", 28.6115, -80.5858)
]

# ------------------------------------------
# Compute closest items
# ------------------------------------------
def find_closest_point(points):
    min_dist = 99999
    closest = None
    for p in points:
        d = haversine(launch_site["lat"], launch_site["lon"], p[1], p[2])
        if d < min_dist:
            min_dist = d
            closest = (p[0], p[1], p[2], round(d, 2))
    return closest

closest_city = find_closest_point(cities)
closest_railway = find_closest_point(railways)
closest_highway = find_closest_point(highways)

# Create a DataFrame
result_df = pd.DataFrame({
    "Launch Site": [launch_site["name"]],
    "Closest City": [closest_city[0]],
    "City Distance (km)": [closest_city[3]],
    "Closest Railway": [closest_railway[0]],
    "Railway Distance (km)": [closest_railway[3]],
    "Closest Highway": [closest_highway[0]],
    "Highway Distance (km)": [closest_highway[3]]
})

print(result_df)

# ------------------------------------------
# Folium Map
# ------------------------------------------
site_map = folium.Map(location=[launch_site["lat"], launch_site["lon"]], zoom_start=10)

# Launch site marker
folium.Marker(
    location=[launch_site["lat"], launch_site["lon"]],
    popup="Launch Site: " + launch_site["name"],
    icon=folium.Icon(color="blue", icon="rocket")
).add_to(site_map)

# --------------------
# Add proximity markers + lines
# --------------------
def add_proximity_marker(name, lat, lon, label):
    folium.Marker(
        location=[lat, lon],
        popup=f"{label}: {name}",
        icon=folium.Icon(color="green")
    ).add_to(site_map)

    folium.PolyLine(
        locations=[[launch_site["lat"], launch_site["lon"]], [lat, lon]],
        weight=2,
        dash_array="5, 5"
    ).add_to(site_map)

add_proximity_marker(closest_city[0], closest_city[1], closest_city[2], "Closest City")
add_proximity_marker(closest_railway[0], closest_railway[1], closest_railway[2], "Closest Railway")
add_proximity_marker(closest_highway[0], closest_highway[1], closest_highway[2], "Closest Highway")

site_map
```
